La siguiente implementación tiene por objetivo demostrar el funcionamiento del algoritmo A\* un algoritmo de busqueda  informada que tiene en cuenta diferentes parametros antes de iniciar su ejecución, este es implementado para hallar, el camino de menor costo entre un par de nodos, siendo un nodo origen ***a*** y uno objetivo o destino ***b***. <br/>
Todo esto es posible siempre y cuando se cumplan unas determinadas condiciones, como la existencia de caminos entre ellos. 
</div>

El algoritmo A\* utiliza una función de evaluación $f(n)=g(n)+h(n)$ , donde $h(n)$ representa el valor heurístico del nodo a evaluar desde el actual, $n$, hasta el final, y $g(n)$, el coste real del camino recorrido para llegar a dicho nodo, $n$, desde el nodo inicial. A\* mantiene dos estructuras de datos auxiliares, que podemos denominar abiertos, implementado como una cola de prioridad (ordenada por el valor $f(n)$ de cada nodo), y cerrados, donde se guarda la información de los nodos que ya han sido visitados. En cada paso del algoritmo, se expande el nodo que esté primero en abiertos, y en caso de que no sea un nodo objetivo, calcula la $f(n)$ de todos sus hijos, los inserta en abiertos, y pasa el nodo evaluado a cerrados.

```python
#Networkx es una libreria para realizar
#NetworkX is a Python language software package for
#the creation, manipulation, and study of the structure, 
#dynamics, and function of complex networks.
import networkx as nx
```

```python
# We are going to use  a Undirected Graph
G=nx.Graph()
#By Default Networkx create a Dictionary of dictionary, 
#in order to have a O(1) access complexity.
G.clear()
```

```python
citynames=['Medellin','Barranquilla','Cartagena','Monteria','Bogota','Neiva','Santa','Marta','Armenia','Pereira','Cali']
```

```python
#G.add_nodes_from(citynames)
for i in citynames : 
    G.add_node(i)
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G)
```

```python
nx.draw_networkx(G)
```

```python
plt.show()
```

```python
G.nodes()
```

```python
for i in G.nodes():
    for j in G.nodes():
        G.add_edge(i,j)

nx.draw_networkx(G)
plt.show()
```
